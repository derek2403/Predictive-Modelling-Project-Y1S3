# Features understanding
**1. Date:** The date of the data entry.

**2. Quarter:** The quarter in which the data was recorded.

**3. Department:** The department to which the data pertains (e.g., sewing).

**4. Day:** The day of the week on which the data was recorded.

**5. Team:** The team associated with the recorded data.

**6. Targeted_Productivity:** The targeted productivity for the given parameters.

**7. SMV:** Standard Minute Value, a measure used in garment production to indicate the time required to complete a task.

**8. WIP:** Work in Progress, referring to the number of garments or units currently being processed.

**9. Over_Time:** The amount of overtime worked.

**10. Incentive:** Any incentives provided to workers.

**11. Idle_Time:** The amount of time that machines or workers were idle.

**12. Idle_Men:** The number of workers who were idle during the recorded time period.

**13. No_of_Style_Change:** The number of style changes that occurred during the recorded time period.

**14. No_of_Workers:** The number of workers involved in the process.

**15. Actual_Productivity:** The actual productivity achieved during the recorded time period.

### Dealing with outliers

```python
import matplotlib.pyplot as plt

num_cols = len(numerical.columns)
num_rows = (num_cols + 1) // 2  # Ensure an even number of rows

fig, axes = plt.subplots(num_rows, 2, figsize=(12, 4*num_rows))

axes = axes.flatten()

for i, col in enumerate(numerical.columns):
    ax = axes[i]
    ax.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

if num_cols % 2 != 0:
    axes[-1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Calculate z-scores for outlier detection
def detect_outliers_z_score(df, col, threshold=3):
    z_scores = (df[col] - df[col].mean()) / df[col].std()
    outliers = df[(np.abs(z_scores) > threshold)]
    return outliers

# Remove outliers using z-score method
def remove_outliers_z_score(df, col, threshold=3):
    z_scores = (df[col] - df[col].mean()) / df[col].std()
    df_no_outliers = df[(np.abs(z_scores) <= threshold)]
    return df_no_outliers
    # Remove outliers
    df = remove_outliers_z_score(df, col)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2]
}

model = GradientBoostingRegressor()

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_

print("Best Parameters:", grid_search.best_params_)
```
